```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel("Years")
plt.ylabel("Yeilds (tons per hectare)")
plt.legend(['apples', 'oranges'])
plt.title("Crop yeilds in Rajshahi");
```

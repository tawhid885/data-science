```python
fig, axis = plt.subplots(2,3, figsize=(16, 8))

#line chart
axis[0,0].plot(years, apples, 'o-r')
axis[0,0].plot(years, oranges, 's--b')
axis[0,0].set_xlabel("Year")
axis[0,0].set_ylabel("Yield (tons per hectare)")
axis[0,0].set_title("Crop yield in Rajshahi")
axis[0,0].legend(['apples', 'oranges'])

#scatter plot
axis[0,1].set_title("Sepal length vs Sepal width")
sns.scatterplot(x='sepal_length', y='sepal_width', hue="species", s=100, ax= axis[0,1], data= flowers_df)

#histogram (multiple)
axis[0,2].set_title("Distributation of Sepal width")
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species=='versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
axis[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins= np.arange(2, 5, 0.25), stacked= True)
axis[0,2].legend(['Setosa', 'versicolor', 'virginica'])

#barplot
axis[1,0].set_title("Restaurant bills")
sns.barplot(x='day', y='total-bill', hue='sex', data=tips_df)

#heatmap
axis[1,1].set_title("No. of passengers (1000's)")
flights_new_df = flights_df.pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights_new_df, annot=True, cmap="Blues", ax= axis[1,1])

#image
axis[1,2].set_title("Image import and showing")
axis[1,2].grid(False)
axis[1,2].set_xticks([])
axis[1,2].set_yticks([])
axis[1,2].imshow(img)

plt.tight_layout(pad=2);
```
